```kotlin
%use @https://raw.githubusercontent.com/gabrielfeo/gradle-enterprise-api-kotlin/6bb80ebf8c46e9a4f5136df1db0b690ba6b07e54/gradle-enterprise-api-kotlin.json
%use lets-plot

import kotlinx.coroutines.*
import kotlinx.coroutines.flow.*
import java.time.*
```

```kotlin
val start = LocalDate.now().minusDays(1).atStartOfDay(ZoneId.of("UTC"))
val builds = runBlocking {
    gradleEnterpriseApi.getGradleAttributesFlow(since = start.toInstant().toEpochMilli()).toList()
}
println(builds.size)
```

```kotlin
val data = mapOf(
    "buildHours" to builds.map {
        Instant.ofEpochMilli(it.buildStartTime).atZone(ZoneId.of("UTC")).hour
    }.sorted()
)
letsPlot(data) + geomHistogram() { x = "buildHours" }
```
